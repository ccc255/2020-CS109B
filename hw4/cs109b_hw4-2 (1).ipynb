Compute this operation by hand assuming a vertical and horizontal stride of 1 as well as a) valid, b) same, and c) full padding modes.

```python
# add_pad function
def add_pad(mat, kernel, form = None):
    kernel_shape = kernel.shape[0]
    if form == None:
        return mat
    else:
        mat = np.concatenate((np.zeros((mat.shape[0], kernel_shape-1)), mat), axis = 1)
        mat = np.concatenate((np.zeros((kernel_shape -1 ,mat.shape[1])), mat), axis = 0)
        if form == 'full':
            mat = np.concatenate((mat, np.zeros((mat.shape[0], kernel_shape - 1))), axis = 1)
            mat = np.concatenate((mat, np.zeros((kernel_shape - 1, mat.shape[1]))), axis = 0)
        return mat
    
# kernel
kernel = np.array([[2,1],[1,3]])

# mats
mat = np.array([[2,3,2,4],
               [3,1,2,2],
               [4,1,0,1],
               [7,2,1,3]])
add = 1
mat_same = add_pad(mat, kernel, 'same')
mat_full = add_pad(mat, kernel, 'full')


# do convolutions 
def conv_nostride(in_matrix, kernel):
    kernel_size_0 = kernel.shape[0]
    kernel_size_1 = kernel.shape[1]
    output_height = in_matrix.shape[0]-kernel_size_0 + 1
    output_width = in_matrix.shape[1]-kernel_size_1 + 1
    out_matrix = np.zeros((output_height, output_width))
    for row in range(output_height):
        for col in range(output_width):
            out_matrix[row,col] = np.sum(kernel*in_matrix[row:row+kernel_size_0, col:col+kernel_size_1])
    return out_matrix

for matrix in [mat, mat_same, mat_full]:
    print(conv_nostride(matrix, kernel), '\n')
```

Pooling operations are often used in convolutional neural networks to reduce the dimensionality of the feature maps as well as overall network complexity. Two main types of pooling are used: AveragePooling and MaxPooling.

Using the matrix below, write the output of the AveragePooling and MaxPooling operations with pool size 2x2 and stride  2x2. Repeat with a stride of 1x1.

$$
\left( \begin{array}{cccc}

```python
# mat
mat = np.array([[1,2,2,4],
               [3,1,2,1],
               [4,1,0,2],
               [5,2,2,1]])

max22 = np.zeros((2,2))
avg22 = np.zeros((2,2))
max21 = np.zeros((3,3))
avg21 = np.zeros((3,3))

def pool(mat_in, mat_out, size, stride, func = 'max'):
    for i, row in enumerate(range(0, mat_in.shape[0]-1, stride)):
        for j, col in enumerate(range(0, mat_in.shape[1]-1, stride)):
            if func == 'avg':
                mat_out[i,j] = np.mean(mat_in[row:row + size, col:col + size])
            else:
                mat_out[i,j] = np.max(mat_in[row:row + size, col:col + size])
    return mat_out

print(pool(mat, max22, 2, 2), '\n')
print(pool(mat, avg22, 2, 2, 'avg'),'\n')
print(pool(mat, max21, 2, 1),'\n')
print(pool(mat, avg21, 2, 1, 'avg'),'\n')
```

```python
def conv_nostride_rbg(tensor, kernel, form = None):
    kernel_size_0 = kernel.shape[0]
    kernel_size_1 = kernel.shape[1]
    matrix_for_size = add_pad(tensor[:,:,0], kernel, form = form)
    output_height = matrix_for_size.shape[0]-kernel_size_0 + 1
    output_width = matrix_for_size.shape[1]-kernel_size_1 + 1
    out_tensor = np.zeros((output_height, output_width, tensor.shape[2]))
    for channel in range(tensor.shape[2]):
        matrix = add_pad(tensor[:,:,channel], kernel, form = form)
        out_tensor[:,:, channel] = conv_nostride(matrix, kernel)
    return np.array(out_tensor)

# filter for sharpening according to https://en.wikipedia.org/wiki/Kernel_(image_processing)
sharpening_filter = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]])

# filter for normalized box blurring according to https://en.wikipedia.org/wiki/Kernel_(image_processing)
normalized_box_blurring_filter = np.array([
    [1/9,1/9,1/9],
    [-1/9,1/9,1/9],
    [1/9,1/9,1/9]])


library_sharp = np.clip(conv_nostride_rbg(library_image_data, sharpening_filter, form = 'same'),0, 1)
library_blurred = np.clip(conv_nostride_rbg(library_image_data, normalized_box_blurring_filter, form = 'same'),0, 1)
```

```python
import certifi
import urllib3  # For handling https certificate verification 
import scipy.ndimage as ndimage
import tensorflow_datasets as tfds
import tensorflow_addons as tfa
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tf_keras_vis.saliency import Saliency
from tf_keras_vis.utils import normalize
from matplotlib import cm
from tf_keras_vis.gradcam import Gradcam

# This line will download the CelebA dataset. The download will only happen the first time you ever run this cell.
train_celeb, test_celeb = tfds.load('celeb_a', split=['train', 'test'], shuffle_files=False)
```

```python
# You may use the following two functions
def normalize_image(img):
    return tf.cast(img, tf.float32)/255.

def rot_resize(img, deg):
    rotimg = ndimage.rotate(img, deg, reshape=False, order=3)
    rotimg = np.clip(rotimg, 0., 1.)
    rotimg = tf.image.resize_with_crop_or_pad(rotimg,140,120)
    return rotimg

################################################################
# Don't manually invoke these functions; they are for Dataset 
# pipelining that is already done for you.
################################################################
def tf_rot_resize(img, deg):
    """Dataset pipe that rotates an image and resizes it to 140x120"""
    rotimg = tfa.image.rotate(img, deg/180.*np.pi, interpolation="BILINEAR")
    rotimg = tf.image.resize_with_crop_or_pad(rotimg,140,120)
    return rotimg

def tf_random_rotate_helper(image):
    """Dataset pipe that normalizes image to [0.,1.] and rotates by a random
    amount of degrees in [-60.,60.], returning an (input,target) pair consisting
    of the rotated and resized image and the degrees it has been rotated by."""
    image = normalize_image(image)
    deg = tf.random.uniform([],-60.,60.)
    return (tf_rot_resize(image,deg), deg)  # (data, label)

def tf_random_rotate_image(element):
    """Given an element drawn from the CelebA dataset, this returns a rotated
    image and the amount it has been rotated by, in degrees."""
    image = element['image']
    image, label = tf_random_rotate_helper(image)
    image.set_shape((140,120,3))
    return image, label
################################################################
```

```python
# Pipeline for creating randomly rotated images with their target labels being 
# the amount they were rotated, in degrees.
train_rot_ds = train_celeb.map(tf_random_rotate_image)
test_rot_ds = test_celeb.map(tf_random_rotate_image)

```

**4.3** Conceptual Question. [2pts] Dropout layers have been shown to work well for regularizing deep neural networks, and can be used for very little computational cost. For our network, is it a good idea to use dropout layers? Explain, being sure to explicitly discuss how a dropout layer works, and what that would mean for our model.


Dropout randomly sets some of the activations to 0 when passing the data between layers. This helps to prevent the network from relying on any given activation too much and it allows the network to learn a representation of the data from an ensemble of activations across layers that are generally not highly correlated to each other (it also helps regularize to prevent overfitting in the setting of having a very high number of parameters). However, it also affects the spatial representation of data since random nodes are set to 0 during training with dropout. For our network, it is therefore not recommended to use dropout because the  data are images and therefore the spatial representation of data and features are important. We can limit the number of parameters during training in order to mitigate overfitting instead of using dropout.

It would work to load the weights and continue training because this is analagous to training a model in a "warm start" situation (where the initial values for the weights are a better fit for the task than random or null values would be). It would be different than continuing from a full-saved model because the weights would continue to be optimized during learning. The final values for the weights learned after restarting training would not necessarily be the same as the final values for the weights saved previously in the fully trained model because the optimization was initiated from different positions in the search space for the weights (and these different initial positions lead to different final weights).
